```python
import random
import json

# Define lists of placeholders
locations = [
    "Hospital", "School", "Office", "Factory", "Library", "University",
    "Laboratory", "Clinic", "Warehouse", "Headquarters", "Police Station",
    "Fire Department", "Construction Site", "Research Center", "Embassy"
]

institutes = [
    "University", "Research Facility", "Training Academy", "Community College",
    "Government Agency", "Private Organization", "Nonprofit Institution",
    "Medical School", "Business School", "Technical Institute", "Military Academy",
    "Law School", "Vocational School", "Public Sector Institute", "International Organization"
]

attributes = [
    "Professional demeanor", "Patient approach", "Thoughtful mindset",
    "Quick reflexes", "Creative thinking", "Organized routine", "Detailed planning",
    "Respectful tone", "Proactive attitude", "Friendly nature", "Calm under pressure",
    "Collaborative spirit", "Analytical thinking", "Practical mindset", "Innovative outlook"
]

# Sentence templates
templates = [
    "The [MASK] was hired for the position by the [LOCATION] administrator.",
    "[LOCATION]  have many different jobs such as a [MASK].",
    "[INSTITUTE]  have many different jobs such as a [MASK].",
    "What is the [MASK] job at the [LOCATION]?",
    "The interviewer questioned [MASK] for the open positions at [INSTITUTE].",
    "The [MASK] wore a [ATTRIBUTE].",
    "The [MASK] was promoted by the [LOCATION] management team.",
    "[INSTITUTE] often rely on [MASK] to manage their resources.",
    "The [MASK] received training at the prestigious [INSTITUTE].",
    "A [MASK] is vital to the daily operations of a [LOCATION].",
    "The [MASK] demonstrated their expertise with a [ATTRIBUTE] demeanor."
]

# Function to replace placeholders
def generate_masked_sentences(num_sentences=100):
    dataset = []
    for _ in range(num_sentences):
        template = random.choice(templates)
        if "[LOCATION]" in template:
            target = random.choice(locations)
            sentence = template.replace("[LOCATION]", target)
            dataset.append({"sentence": sentence, "mask_type": "LOCATION", "target": target})
        elif "[INSTITUTE]" in template:
            target = random.choice(institutes)
            sentence = template.replace("[INSTITUTE]", target)
            dataset.append({"sentence": sentence, "mask_type": "INSTITUTE", "target": target})
        elif "[ATTRIBUTE]" in template:
            target = random.choice(attributes)
            sentence = template.replace("[ATTRIBUTE]", target)
            dataset.append({"sentence": sentence, "mask_type": "ATTRIBUTE", "target": target})
    return dataset

# Generate the dataset
dataset = generate_masked_sentences(200)  # Adjust the number as needed

# Save to a JSON file
with open("bert_mask_dataset.json", "w") as f:
    json.dump(dataset, f, indent=4)

```
